```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    'color':['red','blue','green','blue','red'],
    'shape':['circle','square','Triangle','circle','square']
})
```

```python
print("Original data: \n\n",df)
```

```python
# Label encoding
```

```python
le_color = LabelEncoder()
le_shape = LabelEncoder()
```

```python
df['color_encoded'] = le_color.fit_transform(df['color'])
df['shape_encoded'] = le_shape.fit_transform(df['shape'])
print("\n Label Encoded Data: ")
print(df)
```

```python
 # One Hot Encoding
```

```python
df_encoded_pandas = pd.get_dummies(df, columns=['color','shape'])
print("\n One hot Encoded Data :")
print(df_encoded_pandas)
# df_encoded_pandas
```

```python
# Ordinal Encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size':['small','medium','large','medium','small'],
    'education':['primary','secondary','graduate','post graduate','graduate']
})
df
```

```python
size_order = ['small','medium','large']
education_order = ['primary','secondary','graduate','post graduate']
```

```python
encoder = OrdinalEncoder(categories  = [size_order,education_order])
df_encoded = df.copy()
df_encoded[['size_encoded','education_encoded']] = encoder.fit_transform(df[['size','education']])
print("\n\n Ordinal Encoded Data")
print(df_encoded)
```
