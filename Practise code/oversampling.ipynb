## oversampleing

```python
import pandas as pd
from sklearn.utils import resample
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HP\Desktop\Cybrom work\ML\Practise code\card_transdata.csv")
```

## technique to check if dataset is imbalanced or not

```python
df["fraud"].value_counts(normalize=True) * 100
```

```python
from collections import Counter

print(Counter(df["fraud"]))
```

```python
sns.countplot(x=df['fraud'])
```

```python
df_majority = df[df['fraud']==0]
df_minority = df[df['fraud']==1]
```

```python
df_minority_oversampled = resample(df_minority,replace=True,n_samples= len(df_majority), random_state=42)
```

```python
Oversampled_data = pd.concat([df_majority,df_minority_oversampled])
```

```python
print(Counter(Oversampled_data['fraud']))
```

```python
sns.countplot(x=Oversampled_data['fraud'])
```
