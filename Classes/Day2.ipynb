                  Imputing the data(Filling the nan vales with Mean or Median or Mode)

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.DataFrame({'S.no':[1,2,3,4,5,6],'age':[23,np.nan,34,42,np.nan,26],
'salary':[10000,20000,np.nan,50000,np.nan,45000],'city':["Bhopal","Indore",np.nan,"Ujjain",np.nan,np.nan]})
display(df)
```

                                      Mean imputer

```python
mean_imputer = SimpleImputer(strategy='mean')
df_mean = df.copy()
df_mean[["age","salary"]] = mean_imputer.fit_transform(df[["age","salary"]])
print("\n\n Imputed Data is here")
display(df_mean)
```

                                    Median Imputer

```python
median_imputer = SimpleImputer(strategy = "median")
df_median = df.copy()
df[["age","salary"]] = median_imputer.fit_transform(df[["age","salary"]])
print("\n\n Imputed Data is here")
display(df_median)
```

                                 Mode Imputer

```python
mode_imputer = SimpleImputer(strategy='most_frequent')
df_mode = df.copy()
df[["city"]] = mode_imputer.fit_transform(df[["city"]])
print("\n\n Imputed Data is here")
display(df_mode)
```
