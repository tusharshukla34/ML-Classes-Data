## Encoding Technique
- Label Encoding
- Ordinal Encoding
- One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    'fruits':['Apple','Orange','Apple','Peach','orange'],
    'color':['Red','Blue','Red','Blue','yellow']
})
display(df)
```

## Lable Encoder

```python
le_fruits = LabelEncoder()
le_color = LabelEncoder()
```

```python
df['Encoded_fruits'] = le_fruits.fit_transform(df['fruits'])
df['Encoded_colors'] = le_color.fit_transform(df['color'])
print("\n\n Encoded Data with the help of LabelEncoder")
display(df)
```

## Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df1 = pd.DataFrame({
    'Marks':[50,87,20,34,90],
    'Review':["Good","Best","Poor","Poor","Best"],
    'Effort':["Medium effort","High effort","Poor effort","Poor effort","High effort"]
})
display(df1)
```

```python
review_order = ["Poor","Good","Best"]
result_order = ["Poor effort","Medium effort","High effort"]
```

```python
ordinal_encoder = OrdinalEncoder(categories=[review_order,result_order])
df1[['review_encoded','effort_encoded']] = ordinal_encoder.fit_transform(df1[['Review','Effort']])
print("\n\n Encoded Data with help of Ordinal Encoder")
print(df1)
```

## One Hot Encoding

```python
df_One_hot_encoding = pd.get_dummies(df , columns=['fruits','color'])
print("\n\n Encoded data with help of One Hot Encoder")
df_One_hot_encoding
```
