```python
from sklearn.impute import SimpleImputer
```

```python
# 10-03-2026           --------------- Impute
```

```python
data = {'id':[1,2,3,4,5,6],
        'age':[25,np.nan,30,35,np.nan,100],
        'salary':[50000,60000,np.nan,80000,90000,np.nan],
        'city':["new york","London",np.nan,"paris","London",np.nan]}

df = pd.DataFrame(data)
print("original data:")
display(df)
```

```python
# mean imputer
```

```python
mean_imputer = SimpleImputer(strategy = 'mean')
df_mean = df.copy()
df_mean[["age","salary"]] = mean_imputer.fit_transform(df_mean[["age","salary"]])
print("\n Imputed mean Data:")
display(df_mean)
```

```python
# median imputer
```

```python
median_imputer = SimpleImputer(strategy = "median")
df_median = df.copy()
df_median[["age","salary"]] = median_imputer.fit_transform(df[["age","salary"]])
print("\n Imputed median Data:")
display(df_median)
```

```python
# mode Imputer
```

```python
mode_imputer = SimpleImputer(strategy = "most_frequent")
df_mode = df.copy()
df_mode[["city"]] = mode_imputer.fit_transform(df[["city"]])
print("\n Imputed mode Data:")
display(df_mode)
                             
```

```python
# random imputation
```

```python
df_random = df.copy()
def random_impute(series):
    filled = series.dropna().tolist()
    return series.apply(lambda x: random.choice(filled) if pd.isnull(x) else x)

for col in ['age','salary','city']:
    df_random[col] = random_impute(df_random[col])
    
print("\n Random Imputed Data:")
display(df_random)
```
