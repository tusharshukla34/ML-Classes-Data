```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler, StandardScaler , RobustScaler
```

```python
df = pd.DataFrame({'age':[18,22,30,45,60,80],'salary':[20000,25000,40000,60000,120000,1000000]})
print("Original Data")
display(df)
```

```python
# MinMax Scaler
```

```python
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(df), columns = df.columns)
print("\n Min Max Scalled Data:")
print(df_minmax)
```
