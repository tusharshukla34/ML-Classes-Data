```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
X = np.array([
    [1,7],  #A
    [2,8],  #B
    [3,7],  #C
    [4,6],  #D
    [5,5],  #E
])

y_labels = np.array(['F','F','P','P','P'])

le = LabelEncoder()
y = le.fit_transform(y_labels)
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X,y)
```

```python
new_student = np.array([[3,6.5]])
prediction = knn.predict(new_student)
print("Prediction",le.inverse_transform(prediction)[0])
```

# Create Grid

```python
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

# mesh grid

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),np.arange(y_min, y_max, 0.1))
```

```python
# predict value for each point in the grid
```

```python
Z = knn.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# giving color

cmap_light = ListedColormap(['#FFAAAA','#AAFFAA'])
cmap_bold = {0:'red', 1:'green'}

plt.contour(xx,yy,Z, alpha=0.3,cmap=cmap_light)  # boundary
plt.show
```

```python
for class_value in np.unique(y):
    plt.scatter(X[y==class_value,0], X[y==class_value,1],
    c=cmap_bold[class_value],
    label = le.inverse_transform([class_value])[0],edgecolor = 'k', s=100)
```
